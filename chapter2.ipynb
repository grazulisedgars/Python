```python
# Homework with for loop

initial_investment = 1000
capitalization_goal = 2000


for i in range(2,9) :
    current_investment = initial_investment
    years = 0
    interest_rate = i / 100
    while current_investment < capitalization_goal :
        current_investment = current_investment + current_investment *  interest_rate
        years = years + 1   
    print(f"It will take {years} years to reach capitalization goal ${capitalization_goal} with {interest_rate} interest rate")
```

```python
# while loop practice (while loops until a condition is met)

initial_wallet = 1000
interest_rate = 0.05

years = 0 
current_wallet = initial_wallet

while current_wallet < 2* initial_wallet :
    current_wallet = current_wallet + current_wallet * interest_rate # current_wallet = current_wallet * (1 + interest_rate)
    years = years +1 # years +=1
    print('My investment is worth ', current_wallet, 'on year ', years)
```

```python
# Variables
savings_goal = 10000
monthly_savings = 400
month_of_raise = 6
raise_amount = 200

# Counter initialization
months_required = 0
total_savings = 0

# Loop until you reach the savings goal
while total_savings < savings_goal :
    # Check if the raise is in effect and adjust monthly savings
    if months_required == month_of_raise :
        monthly_savings = monthly_savings + raise_amount
        # print("I got a raise")
        
    # Update total savings
    total_savings = total_savings + monthly_savings
    # print(f"On month {months_required +1} you have {total_savings}")
    
    # Increment the months counter
    months_required = months_required + 1
    
print(f"You will need {months_required} months to reach our savings goal of {savings_goal}")
```
