```python
data = {
    "date": ["2022-01-01", "2022-01-02", "2022-01-03", "2022-01-04"],
    "close": [150.0, 152.0, 153.0, 154.0],
    "volume": [1000, 1200, 1300, 1250],
}

df = pd.DataFrame(data)
#print(df)
df # Because we are in jupyter notebook calling this give better output
```

```python
print(df.index) # can access index
print(list(df.index)) # can print them as lists

#print(df['volume'])
print(df.loc[2]) # We access 2nd row
print(df.loc[2]['volume']) # We access specific value
print(df.loc[2][['volume', 'close']]) # We can also access in such way using a list
print(df[['volume','close']]) # this prints columns
```

```python
df['volume'] > 1300 # Gives output false as nothing in volume column is larger than 1300
```

```python
df.loc[df['volume'] >= 1250]
df.loc[(df['volume'] >= 1250) & (df['close'] > 153)]
df.loc[(df['volume'] >= 1250) | (df['close'] > 153)] # | symbol means "or"
df['volume'].loc[(df['volume'] >= 1300) | (df['close'] >153)] # We have isolated only volume part 
```

```python
df["time"] = pd.to_datetime(df["time"], unit='s') # Take the column time from dataframe and change it to date time. Then inside method we state that df['time] column is in seconds
print(df["time"])
```

```python
df['open'] ==0 #not helpful because the 0 is somewhere in the middle of the data set.
df.loc[df['open'] == 0] # locating specific row
```

```python
df.set_index('time', inplace=True)
df
```

```python
df.loc['2023-03-13'] # Specific day
df.loc['2023-03'] # a month of march
df.loc['2023'] # year of 2023 data 
df.loc['2021':] # from 2021 till today
df.loc['2021': '2022'] # in between those years
```

```python
df.iloc[3:10]
#del df['close']
```

```python
df['low_high_avg'] = (df['low'] + df['high']) / 2 # This one line has calculated whole column with values. This is also better than to make a loop and calculate line by line
df
```

```python
df.loc[df['open']==0] # We again locate the row, column where we have 0 as value
```

```python
df.loc[df['open'] == 0, 'open'] = df['low_high_avg'] # we have replaced that 0 value with avg value
df.loc[df['open'] ==0] # We can no longer locate value that is 0
```

```python
df.loc['2015-10-05'] # We see that for the same row open value has been replaced and is no longer 0 
```

```python
mean_prices = df['close'].mean()
median_prices = df['close'].median()
std_prices = df['close'].std()

print("Average closing prices:", round(mean_prices,2))
print("Median closing prices:",round(median_prices,2))
print("Standard deviation of closing prices:", round(std_prices,2))
```

```python
df['moving_average'] = df['close'].rolling(20).mean() #Rolling(2) means that moving average takes last two value of close. That is why first is NaN
df
```

```python
df['dayly_returns'] = df['close'].pct_change()
df['volatility'] = df['dayly_returns'].rolling(30).std()
df
```

```python
df.sort_values(by='dayly_returns', ascending=True, inplace=True)
df['change'] = (df['close'] - df['open']) / df['close']
df.sort_values(by='change', ascending=True, inplace=True)
df
```
