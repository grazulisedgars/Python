```python
# Function calculates compound interest, build up on previous interest

def compound_interest(principal, rate, time): # principal initial ammount of investment, rate is the interest rate, time in years
    amount = principal * (1 + rate) ** time # Amount will be what is after = ** means power (exponentiation) 2**3 = 8 meaning 2*2*2
    interest = amount - principal # with this line we define the interest that we earned
    return interest # this lines returns the result what function above calculates

a =compound_interest(1000, 0.05,10) # We assign function to variable a 
print(a) # we print function via variable a
```

```python
def compound_interest(principal, rate, time):
    amount = principal * (1+rate) ** time
    interest = amount - principal
    return interest

# Defining variables with values
initial_investment = 1000
interest_rate = 0.04
years = 5

interest = compound_interest(initial_investment, interest_rate, years) # Calling variables values we defined earlier
print(f"The compound interest after {years} years is {round(interest,2)}$") # f is f string notation shortens and helps make things clearer. New round functions, we call function and the result is shortened to 2 decimals after comma
```

```python
class BankAccount: # Creating a class bank account
    def __init__(self, balance): # special method (function inside class), self refer to specific object e.g. account1, balance is a parameter we pass
        print(f'Initialization with {balance} USD') # prints balance
        self.balance = balance # stores the balance inside object
    
    def withdraw(self, amount): #defines method (function inside a class) takes self the object and amount how much to withdraw
        print(f'Withdraw of {amount}') #prints how much is withdrawn
        self.balance = self.balance - amount # Updates the balance by subtracting amount
        
account1 = BankAccount(10000) # This is the object that we create, 10000 is passed to __init__
account1.withdraw(2000) # Calls the withdraw method(function)
print(account1.balance) #Accesses the balance stored in the object and gives output
```

```python
class BankAccount:
    def __init__(self, balance = 0): # Constructor
        self.balance = balance
        
    def deposit(self,amount):
        self.balance = self.balance + amount
        print(f'You have deposited {amount} $. New balance {self.balance} $')
    
    def withdraw(self,amount):
        if amount > self.balance:
            print('Insufficient balance')
        else:
            self.balance = self.balance - amount
            print(f"You've withdrawn {amount} $. New balance: {self.balance} $")
    
    def display_balance(self):
        print(f'Your balance is {self.balance}$')
        
account1 = BankAccount()
account1.deposit(100)
account1.withdraw(120)
account1.display_balance()
```
