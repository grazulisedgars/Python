```python
class BankAccount:
    def __init__(self, balance = 0): # Constructor
        self.balance = balance
        
    def deposit(self,amount):
        self.balance = self.balance + amount
        print(f'You have deposited {amount} $. New balance {self.balance} $')
    
    def withdraw(self,amount):
        if amount > self.balance:
            print('Insufficient balance')
        else:
            self.balance = self.balance - amount
            print(f"You've withdrawn {amount} $. New balance: {self.balance} $")
    
    def display_balance(self):
        print(f'Your balance is {self.balance}$')
        
account1 = BankAccount()
account1.deposit(100)
account1.withdraw(120)
account1.display_balance()
```

```python
# Homework

class Portfolio:
    def __init__(self): # self represents the object, the whole class
        self.assets = {} # creating an empty dictionary
    
    def add_asset(self, name, quantity, unit_value): # this method will allow to add a new asset or update existing one when calling my_portfolio.add_asset("AAPL", 10, 150)
        if name in self.assets: # we check if we already have this asset. I guess we refer back here to the empty dictionary we made above
            # update existing asset
            self.assets[name]['quantity'] = self.assets[name]['quantity'] + quantity # or we can also write it as self.assets[name]['quantity'] += quantity
            self.assets[name]['unit_value'] = unit_value # updates price to the latest value
        else: # this runs if we are adding asset for the first time
            # create new asset
            self.assets[name] = { #creates new dictionary entry. So self.assets is an empty dictionary we established at the top. This line means go into the dictionary assets and accessthe value stored aunder the key "name"
                "quantity": quantity,
                "unit_value": unit_value
            }
    
    def remove_asset(self, name):
        if name not in self.assets:
            print('Asset does not exist in the portfolio')
        else:
            print(f"We have removed {name}")
            del self.assets[name]
            
    
    def update_asset(self, name, quantity, unit_value):
        if name not in self.assets:
            print('Asset is not in portfolio')
        else:
            self.assets[name]["quantity"] = quantity
            self.assets[name]["unit_value"] = unit_value
            print(f'You have updated asset {name} quantity to {quantity} and value {unit_value}')
            
    def get_total_value(self):
        total_value = 0 # creates a starting point, accumulator. Think of it as empty bucket
        for key in self.assets:
            total_value = total_value + self.assets[key]["quantity"] * self.assets[key]["unit_value"]
        return total_value
    
    def __str__(self):
        return str(self.assets)
        
my_portfolio = Portfolio()

my_portfolio.add_asset("AAPL", 10, 150) # We call method which passes values into def add_asset(self, name, quantity, unit_value). So now inside function it is def add_asset(self, "AAPL", 10, 150). Then it goes to self.assets[name] = {} and subsitutes those values
# We are not passing information straight into dictionary. We pass values into a function. The function then uses values to update the dictionary
my_portfolio.add_asset("GOOG", 5, 1000)
my_portfolio.add_asset("TSLA", 8, 800)

my_portfolio.update_asset("GOOG", 7, 1500)
my_portfolio.update_asset("KIKO", 2, 2)

my_portfolio.remove_asset("AAPL")
my_portfolio.remove_asset("IBKR")

total_value = my_portfolio.get_total_value()
print(f"The total portfolio value is {total_value}$.")
```
