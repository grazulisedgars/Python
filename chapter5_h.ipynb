```python
df.loc[df['Age'] < 1] # Locates persons who are infants. Just checking if I should include persons with 0 years
df['Age'].isnull()
no_age_data = df['Age'].dropna() # Removing all NaN from Age column
no_age_data.isnull().sum()

# Calculate the average age of passengers
avg_age = no_age_data.mean()
print("The average age is", round(avg_age,2))
```

```python
# What this method does

df.describe()

# 1. Count - Returns number of non-missing (NaN) values. Important: tells you about missing data.
# 2. Mean - Average value. For example average age among passengers 29.7 years
# 3. std - standard deviation. How spread out the data is. Low std means data is tightly grouped. High std data is spread out.
# Group A (low std) - [28,29,30,31,32] values are very close to each other.
# Group B (high std) - [5.15,30,50,80] values are spread out. In titanic example std is 14 which means that ages vary a lot. Have babies nad elderly people
# 4. min - Smallest value (youngest passenger, cheapest fare in this case)
# 5. 25% - first quartile 25% of data is below this value
# 6. 50% (median) - Middle value (not average) Example [10,20,30,40,50] Median = middle number is 30
# 7. 75% - third quartile 75% of data is below this value. Helps understand upper range
# 8. max - Largest value (oldest passenger or highest ticket price)
```

```python
# Identify the oldest and youngest passangers by name
youngest = df['Age'].min() # youngest is just a number we extract from data frame
oldest = df['Age'].max()

youngest_passengers = df[df['Age'] == youngest]['Name']
# df['Age'] == youngest in background creates [False, False, True, False] asking is this row's Age equal to the youngest age?
# df[df'Age']== youngest] keeps only rows where the conditions are true. We get only rows where Age == youngest
# df[df['Age] == youngest]['Name'] from those filtered rows take only "Name" column
oldest_passangers = df[df['Age'] == oldest]['Name']


print("Youngest passenger(s):")
print(youngest_passengers)

print("\nOldest passenger(s):")
print(oldest_passangers)
```

```python
# Determine how many passangers survived and how many perished
survived = 1
died = 0

survived_passengers = df[df['Survived'] == survived]
died_passengers = df[df['Survived'] == died]

print(f"Total of passangers who survived: {len(survived_passengers)}")
print(f"Total of passangers who died: {len(died_passengers)}")
```

```python
# Count the number of male and female passengers aboard the Titanic

male = "male"
female = "female"

male_on_board = df[df['Sex'] == male]
female_on_board = df[df['Sex'] == female]

print(f"Titanic had {len(male_on_board)} male passangers")
print(f"Titanic had {len(female_on_board)} female passangers")

df['Sex'].value_counts() # More panda way to do the same task. One line vs few lines above
```

```python
# How many man and women survived

# With below line from the dataframe keep only rows where sex is male and survived is 1
survived_men = df[(df['Sex'] == 'male') & (df['Survived'] == 1)]
survived_female = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]

print(f"Out of all survivors {len(survived_men)} were men")
print(f"Out of all survivors {len(survived_female)} were women")
```

```python
# What's the survival rate by class? 

# df['Pclass] == 3 filters class 3
# ['Survived'] take survival column
# .mean() average of 0/1 = survival rate

rate_3 = df[df['Pclass'] == 3]['Survived'].mean()
print(f"Survival rate among 3rd class is {round(rate_3,2)}")

# Better solution below
all_rates = df.groupby('Pclass')['Survived'].mean()
print(f"Survival rate: {round(all_rates,2)}")
```

```python
# Add a new 'FamilySize' column that sums the 'SibSp' (number of siblings/spouses aboard) and 'Parch' (number of parents/ children aboard) columns
df['FamilySize'] = df['SibSp'] + df['Parch']
df
```

```python
# Identify the non-survivor with the most significant family on board

non_survivors = df[df['Survived'] ==0]
max_family_size = non_survivors['FamilySize'].max()
result = non_survivors[non_survivors['FamilySize'] == max_family_size]

print(result['Name'])  
```
