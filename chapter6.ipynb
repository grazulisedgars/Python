```python
import matplotlib.pyplot as plt
```

```python
# Line Plot
x= [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Example Plot')
plt.show()
```

```python
# Bar Plot

x= [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.bar(x,y)
plt.show()
```

```python
# Scatter Plot

x= [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.scatter(x,y)
plt.show()
```
