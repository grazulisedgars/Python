```python
import matplotlib.pyplot as plt
```

```python
# Line Plot
x= [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Example Plot')
plt.show()
```

```python
# Bar Plot

x= [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.bar(x,y)
plt.show()
```

```python
# Scatter Plot

x= [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.scatter(x,y)
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('house_sales_in_king_county_usa.csv')
df
```

```python
df['price'].mean() # average house price
```

```python
# Histograms
plt.hist(df['price'], bins = 100)
plt.xlabel('Price')
plt.ylabel('Nr houses')
plt.show()

# Video till 13:43 min
```

```python
# df['price'] - selects price column
# df['price'] <2e6 - creates boolean condition 2e6 = 2x 10^6 = 2 000 000
# so this means df['price'] < 2 000 000 keep prices below 2 million
# df.loc[] filters rows using the condition
# df.loc[df['price'] < 2e6] give me only rows where price is below 2 million
# ['price'] After filtering rows, you select only the price column
# plt.hist(...) creates a histogram. Histogram groups values into ranges and counts how many values fall into each range
# bins =20 This controls how many groups/buckets the histogram uses. Prices are split into 20 ranges. Bar height how many houses fall into that range

plt.hist(df.loc[df['price'] < 2e6]['price'], bins=20)
plt.xlabel('price')
plt.ylabel('Nr houses')
plt.show()
```

```python
df6 = df.loc[df['bedrooms'] <7]


plt.figure(figsize= (15, 8))
scatter = plt.scatter(df6['sqft_living'], df6['price'], s=5, c=df6['bedrooms'], cmap='inferno')
plt.title('Price as a function of the living area')
plt.xlabel('Living Area')
plt.ylabel('Price')

cbar = plt.colorbar(scatter)
cbar.set_label('N bedroom')
plt.show()

# Till 24:42 min
```
