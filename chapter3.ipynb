```python
# Lists

my_list = [1,2,4]
print(my_list)

my_list_two = ['bonds', 'stocks', 'cryptos'] # Each item in list seperated by comma
print(my_list_two)

my_list_three = ['bonds', 1, 1.3, True, ['another', 'list']] # Kist can contain different types
print(my_list_three)
```

```python
# How to access elements in the list

print(my_list[1]) # for example accessing second element in first list

print(my_list_two[-2]) # we can read list from other end

print (my_list_three[0:3]) # we can read a range of items in list

print (my_list_two[0:-1:2]) # jumps and shows every other item in list

print(my_list_three[-1][0]) # We can access another list within a list
```

```python
# Different things we can do with lists
assets = ['stocks', 'bonds', 'real estate', 'cash']
print(assets)

assets[0] = 'crypto' # we change the first element of a list
print(assets)

assets.append("warrants") # Added extra item at end of list
print(assets)

assets.insert(1,'other security') # We can insert in the list wherever we want
print(assets)

assets.remove('bonds') # We can remove item from a list
print(assets)

del assets[0] # We can delete a list item
print(assets)

print(assets * 3) # We can multiply lists
```

```python
# Loop with Lists
assets = ['Yolo', 'Hello', 'Mello', 'Yello']
for i in range (len(assets)): # This is first option
    print(assets[i])
    
for el in assets : # This is second option el takes value what is in the list
    print(el)
```

```python
# Looping through dictionary
asset_info = {
    'stocks': 'investment in company shares',
    'bonds' : 'investment in debt securities',
    'real estate' : 'investment in real estate',
    'cash' : 'investment in cash or near cash items'
}
for key in asset_info: # Python gives you the keys, not the values
    print(key,asset_info[key]) # This print is divided in two parts, prints key's and asset_info values
    
print(asset_info) # Prints dictionary as it is stated above
print(list(asset_info.keys())) # Prints only keys
print(list(asset_info.values())) # Prints only value 


for value in asset_info.values(): # Loops only through values
    print(value)
```

```python
# Homework 

transactions = [
    {
        "symbol": "BTC",
        "type": "buy",
        "quantity": 0.7,
        "price": 27090.4,
        "date": "2022-03-07",
    },
    {
        "symbol": "BTC",
        "type": "sell",
        "quantity": 0.1,
        "price": 26561.4,
        "date": "2022-04-12",
    },
    {
        "symbol": "BTC",
        "type": "buy",
        "quantity": 0.4,
        "price": 28090.4,
        "date": "2022-05-21",
    },
    {
        "symbol": "BTC",
        "type": "buy",
        "quantity": 0.3,
        "price": 25090.4,
        "date": "2022-06-02",
    },
    {
        "symbol": "BTC",
        "type": "sell",
        "quantity": 0.5,
        "price": 32034.67,
        "date": "2022-06-29",
    },
    {
        "symbol": "ETH",
        "type": "buy",
        "quantity": 6,
        "price": 1754,
        "date": "2022-05-21",
    },
    {
        "symbol": "ETH",
        "type": "buy",
        "quantity": 2,
        "price": 1950,
        "date": "2022-06-02",
    },
    {
        "symbol": "ETH",
        "type": "sell",
        "quantity": 3,
        "price": 2100.67,
        "date": "2022-06-29",
    },
]
```

```python
# 1. Using a loop create a dictionary with keys for each of the cryptocurrencies present

portfolio = {} # We create an empty dictionary

for i in range (len(transactions)): # Each iteration gives you one dictionary from the transactions list
    symbol = transactions[i]["symbol"] #extracts BTC and ETH
    
    if symbol not in portfolio: #check if we have seen it before
        portfolio[symbol] = {} #add it as a key with empty dictionary
   
print(portfolio)
```

```python
# 2. Add to the dictionary a key "quantity" which has as its value the quantity still present in the portfolio: 

for transaction in transactions : # Each iteration gives you one dictionary from list transactions
    symbol = transaction["symbol"] # We define symbol, loop goes through transactions list. We take each symbol value by transaction["symbol"]
    if "quantity" not in portfolio[symbol]:
        portfolio[symbol]["quantity"] = 0 # If quantity is not yet present in portfolio dictionary we set value to start from 0
    if transaction["type"] == "buy": # Transaction loop iterates through transactions list and checks type
        portfolio[symbol]["quantity"] += transaction["quantity"] # If type buy then we update the stored total quantity for that symbol by adding the transaction amount
    elif transaction["type"] == "sell": # If type sell then subtract quantity
        portfolio[symbol]["quantity"]-= transaction["quantity"]
    
print(portfolio) #prints totals in quantity
```
